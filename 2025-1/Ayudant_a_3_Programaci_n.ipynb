## Problema 1
Imprima los números del 1 al n, con n una entrada definida por el usuario.
Opcionalmente puede añadir un delay en la impresión para hacer un cronómetro.

```python
from time import sleep
contador = int(input("Ingrese un número para iniciar el contador: "))
segundo = 0
while segundo <= contador:
    print(segundo)
    segundo += 1
    sleep(1)
```

## Problema 3
El programa debe generar un número aleatorio entre 1 y 100 y el usuario tiene que adivinarlo.

Después de cada intento, el programa debe decir si el número es mayor o menor.

El juego termina cuando el usuario adivina el número.

```python
import random

numero_secreto = random.randint(1, 100)
intentos = 0

print("¡Adivina el número entre 1 y 100!")

flag = True

while flag:
    intento = int(input("Introduce tu intento: "))
    intentos += 1

    if intento < numero_secreto:
        print("Demasiado bajo.")
    elif intento > numero_secreto:
        print("Demasiado alto.")
    else:
        print(f"¡Correcto! Adivinaste en {intentos} intentos.")
        flag = False
```

## Problema 4
Pide al usuario una palabra y cuenta cuántas vocales contiene.

```python
palabra = input("Escribe una palabra: ")
vocales = "aeiouAEIOU"
contador = 0

for letra in palabra:
    if letra in vocales:
        contador += 1

print(f"La palabra tiene {contador} vocal(es).")
```

## Problema 5
Muestra un menú con opciones y ejecuta una acción según lo que el usuario elija. Repite hasta que el usuario seleccione “Salir”.

Las opciones a elegir son:
1. Saludar -> Pregunte el nombre y salude al usuario.
2. Calcular cuadrado de un número -> pida un número y calcule su cuadrado.
3. Salir -> Salga del menú e imprima un mensaje de despedida.

```python
flag=True
while flag:
    print("\nMenú:")
    print("1. Saludar")
    print("2. Calcular cuadrado de un número")
    print("3. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("¿Cómo te llamas? ")
        print(f"¡Hola, {nombre}!")
    elif opcion == "2":
        numero = float(input("Ingresa un número: "))
        print(f"El cuadrado de {numero} es {numero ** 2}")
    elif opcion == "3":
        print("Adiós.")
        flag = False
    else:
        print("Opción no válida.")
```
