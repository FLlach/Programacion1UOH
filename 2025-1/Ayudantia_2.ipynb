##Problema 1

Realice un programa que simule una compra online en una tienda.
Para esto, usted debe solicitar el nombre del Cliente, la cantidad de productos que desea comprar y el precio unitario de los productos.

Calcule el precio total y, si el monto total supera los 50.000 CLP almacene una variable de tipo booleano si el cliente es elegible para recibir el envío gratis.

El output solicitado es del siguiente estilo:

Ingrese el nombre del cliente: Luis Miranda

¿Cuantos productos va a llevar?: 3

¿Cual es el precio unitario de los productos?: 25000

Hola Luis Miranda, el precio total a pagar es 75000 CLP

¿Califica Luis Miranda para envío gratis?: True


```python
# Entrada de datos
nombre = input("Ingrese su nombre: ")
cantidad = int(input("Ingrese la cantidad de productos: "))
precio_unitario = float(input("Ingrese el precio unitario: "))

# Cálculo
total = cantidad * precio_unitario
envio_gratis = total > 50000

# Salida
print(f"\nHola {nombre}, el total a pagar es: ${total}")
print(f"¿Califica {nombre} para envío gratis? {envio_gratis}")
```

##Problema 2

Escriba un programa que calcule el total de una cuenta en un restaurante, para esto,pida al usuario que ingrese el total de la cuenta, pida el porcentaje de propina que desea dejar, calcule la propina y el total a pagar para finalmente mostrarlo en pantalla con un formato correspondiente.

```python
# Entradas como texto
total_cuenta_str = input("Ingrese el total de la cuenta: ")
porcentaje_propina_str = input("Ingrese el porcentaje de propina que desea dejar: ")

# Conversión a tipos numéricos
total_cuenta = int(total_cuenta_str)
porcentaje_propina = int(porcentaje_propina_str)

# Cálculos
propina = total_cuenta * (porcentaje_propina / 100)
total_final = total_cuenta + propina

# Resultados
print(f"\nPropina: ${propina}")
print(f"Total a pagar: ${total_final}")
```
