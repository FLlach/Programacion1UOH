# Repaso de Estructuras de Datos en Python

En este notebook revisaremos las cuatro estructuras de datos integradas más comunes en Python: **Listas**, **Tuplas**, **Sets (Conjuntos)** y **Diccionarios**.

## 1. Listas (`list`)

**Definición:** Una lista es una colección **ordenada** y **mutable** de elementos. Permite duplicados y sus elementos pueden ser de diferentes tipos. Se definen utilizando corchetes `[]`.

```python
# Demostración de Listas
mi_lista = ["Python", 10, 3.14, True, "Python"]

print(f"Lista completa: {mi_lista}")
print(f"Primer elemento: {mi_lista[0]}")

# Las listas son mutables (podemos cambiar elementos)
mi_lista[1] = 20
mi_lista.append("Nuevo elemento")

print(f"Lista modificada: {mi_lista}")
```

```python
numeros = [10, 20, 30, 40, 50]
numeros.append(60)
numeros.pop(1) # Elimina el segundo elemento (índice 1)
print(f"Solución Listas: {numeros}")
```

## 2. Tuplas (`tuple`)

**Definición:** Una tupla es una colección **ordenada** e **inmutable**. Al igual que las listas, permite duplicados, pero una vez creada, no se puede modificar (no se pueden agregar, eliminar o cambiar elementos). Se definen utilizando paréntesis `()`.

```python
# Demostración de Tuplas
mi_tupla = ("Lunes", "Martes", "Miércoles", "Lunes")

print(f"Tupla completa: {mi_tupla}")
print(f"Frecuencia de 'Lunes': {mi_tupla.count('Lunes')}")

# Intentar modificar una tupla daría error:
# mi_tupla[0] = "Domingo"  # <-- TypeError
```

## 3. Sets o Conjuntos (`set`)

**Definición:** Un set es una colección **desordenada**, **no indexada** y **no permite elementos duplicados**. Son muy útiles para realizar operaciones matemáticas de conjuntos como unión o intersección. Se definen utilizando llaves `{}`.

```python
# Demostración de Sets
mi_set = {"Manzana", "Banana", "Cereza", "Manzana"}

print(f"Set (sin duplicados): {mi_set}")

# Operaciones de conjuntos
otro_set = {"Cereza", "Piña", "Uva"}
print(f"Intersección: {mi_set.intersection(otro_set)}")
print(f"Unión: {mi_set.union(otro_set)}")
```

### Ejercicio de Sets
**Enunciado:** Crea dos sets: uno con los números del 1 al 5 y otro con los números del 4 al 8. Encuentra la unión y la diferencia entre ambos conjuntos.

```python
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
print(f"Unión: {set_a.union(set_b)}")
print(f"Diferencia (a - b): {set_a.difference(set_b)}")
```

## 4. Diccionarios (`dict`)

**Definición:** Un diccionario es una colección **ordenada** (desde Python 3.7+), **mutable** y **mapeada**. Almacena datos en pares de `clave: valor`. Las claves deben ser únicas e inmutables. Se definen utilizando llaves `{}` con dos puntos entre clave y valor.

```python
# Demostración de Diccionarios
mi_diccionario = {
    "nombre": "Juan",
    "edad": 25,
    "ciudad": "Santiago"
}

print(f"Diccionario: {mi_diccionario}")
print(f"Edad de Juan: {mi_diccionario['edad']}")

# Modificar y agregar
mi_diccionario["edad"] = 26
mi_diccionario["profesion"] = "Ingeniero"

print(f"Diccionario actualizado: {mi_diccionario}")
```

### Ejercicio de Diccionarios
**Enunciado:** Crea un diccionario que represente un libro (título, autor, año). Agrega una nueva clave para el número de páginas y elimina la clave del año de publicación.

```python
libro = {
    "título": "Don Quijote",
    "autor": "Miguel de Cervantes",
    "año": 1605
}
libro["páginas"] = 1032
del libro["año"]
print(f"Solución Diccionario: {libro}")
```
